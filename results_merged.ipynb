### Q5: For all of the models that were produced in 2008 that are still being produced now, how much has the mpg improved and which vehicle improved the most?

```python
df_modelmpg8 = df.groupby('model_2008').mean().cmb_mpg_2008
df_modelmpg8
```

```python
df_modelmpg18 = df.groupby('model').mean().cmb_mpg
df_modelmpg18
```

```python
df['mpg_change'] = df_modelmpg8.sub(df_modelmpg18, fill_value=0)
df.head()
df['mpg_change']
```
